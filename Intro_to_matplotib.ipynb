# Matplotlib example workflow

```python
#0. import matplotib and get it ready for plotting in jupyter
%matplotlib inline
import matplotlib.pyplot as plt

#1. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(5, 10)) #(width, height)

# 3. Plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title="Simple Plot", 
       xlabel= "x-axis", 
       ylabel="y-axis")

#5. Save & show (you save the whole figure)
fig.savefig("images/sample-plot.png")
```

# Two option for subplots

```python
#Subplot opt. 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10, 5))
#Plot to each different axis
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10), np.random.random(10));                              
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());  
ax4.hist(np.random.randn(1000));
```

```python
#Subplot opt.2
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10,5))
# Plot to each different index
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));                              
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values());  
ax[1,1].hist(np.random.randn(1000));
```

```python
series = pd.Series([63,78,58,56,57,64,82])
series
colours = pd.Series([1,1,0,1,0,0,0])
colours
cp = pd.Series([3,2,1,1,0,1,2])
cp
Odometer = pd.Series([145,130,130,120,120,120,120])
Odometer
Door = pd.Series([233,250,204,236,254,345,543])
Door
Price = pd.Series([1,0,0,0,0,1,1])
Price
Date = pd.Series([0,1,0,1,1,1,0])
Date
Total = pd.Series([150,187,172,178,163,256,651])
Total
exang = pd.Series([0,0,0,0,1,1,1])
exang
oldpeak = pd.Series([2.3,3.5,1.4,0.8,0.6,0.1,0.4])
oldpeak
slope = pd.Series([0,0,2,2,2,0,2])
slope
ca = pd.Series([0,0,0,0,0,0,0])
ca
thal = pd.Series([1,2,2,2,2,2,2])
thal
target = pd.Series([1,1,1,1,1,1,1])
target
heart_disease = pd.DataFrame({"age":series, "sex": colours,"cp":cp,"trestbps":Odometer,"chol":Door,"fbs":Price,"restecg":Date,"thalach":Total,"exang":exang,"oldpeak":oldpeak,"slope":slope,"ca":ca,"thal":thal,"target":target})
heart_disease
```

### Which one shopuld you use? (pyplot vs matplotlib Old method)
*When plotting something quickly,okay to use the pyplot method
*When plotting something more advanced, use the Old method 

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50
```

```python
#pyplot method
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
            figsize=(10,6));
```

```python
#Old method ,object oriented method
fix, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);

#ax.set_xlim([45,100])  to adjust limitations of your access
```

```python
##OO method from scrach

fig, ax = plt.subplots(figsize=(10, 6))

#plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                    cmap="winter") #this chandes colour

#Customize plot

ax.set(title="Heart Disease and Cholestrol Levels",
       xlabel="Age",
       ylabel="Cholesterol");


#Add a legend
ax.legend(*scatter.legend_elements(),title="Target");

# Add horizontal line
ax.axhline(over_50["chol"].mean(),
           linestyle= '--');
```

```python
#Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10))

#Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="winter")

#Customize ax0
ax0.set(title="Heart Disease and Cholestrol Levels",
       xlabel="Age",
       ylabel="Cholesterol");

#Add legend to ax0
ax0.legend(*scatter.legend_elements(),title="Target");

# Add horizontal line
ax0.axhline(over_50["chol"].mean(),
           linestyle= '--');
#Add data to ax1

scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="winter")

#Customize ax1
ax1.set(title="Heart Disease and Cholestrol Levels",
       xlabel="Age",
       ylabel="Cholesterol");

#Add legend to ax1
ax1.legend(*scatter.legend_elements(),title="Target");

# Add horizontal line
ax1.axhline(over_50["chol"].mean(),
           linestyle= '--');

#Add a title to a figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight='bold');
```
